In this notebook, three classes are defined. `env` contains code necessary for creating acoustic environments, and `monopole` and `planeWave` are two source classes, which can be added into an `env` using the `addSource` method. 

Firstly, setting up the notebook:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation, rc
from IPython.display import HTML
from scipy.interpolate import RectBivariateSpline
plt.rcParams['figure.figsize'] = [10, 5]
```

And now, defining the classes

```python
class env(object):
    """ 2D environment

    METHODS
    -------
    __init__ : initialise internal variables
    pressure : plot pressure amplitude
    displacement : plot particle displacement
    spl : plot sound pressure level
    addSource : add a source into the environment """

    def __init__(self, xlim, ylim, resolution, freq, c=343, rho=1.225):
        """ Initialise internal variables
        INPUTS
        ------
        x = 2-tuple (xmin, xmax)
        y = 2-tuple (ymin, ymax)
        resolution = scalar number of points per linear metre
        freq = scalar frequency of harmonic sources
        c = scalar speed of sound - defaults to 343 m/s
        rho = scalar fluid density - defaults to standard air 1.225 kg/m^3
        """
        
        assert len(list(xlim)) == 2, 'Usage: xlim = [xmin, xmax]'
        assert len(list(ylim)) == 2, 'Usage: ylim = [ymin, ymax]'
        xmin, xmax = xlim
        ymin, ymax = ylim
        x = np.linspace(xmin, xmax, np.int(np.floor((xmax-xmin)*resolution)))
        y = np.linspace(ymin, ymax, np.int(np.floor((ymax-ymin)*resolution)))
        self.x = x
        self.y = y
        self.X, self.Y = np.meshgrid(x, y)  # set X and Y coordinate grids
        self.p = np.zeros_like(self.X, dtype=np.complex128)  # pressure grid
        self.gradpx = np.zeros_like(self.X, dtype=np.complex128) # pressure gradient in x direction
        self.gradpy = np.zeros_like(self.X, dtype=np.complex128) # pressure gradient in y direction
        self.sourceList = []
        self.freq = freq
        self.c = c
        self.rho = rho
        
    def pressure(self,clim=(-2,2)):
        """ Plot pressure in the environment at t = 0 
        INPUTS 
        ------
        Optional clim 2-tuple (lower colour limit, upper colour limit)
        """
        fig,ax = plt.subplots()
        ax.extent = (self.X.min(), self.X.max(), self.Y.min(), self.Y.max())
        surface = ax.pcolormesh(self.X,self.Y,np.real(self.p),vmin=clim[0],vmax=clim[1])
        fig.colorbar(surface)
        plt.title('Pressure Amplitude (Pa)')
        plt.axis('equal')

    def spl(self,clim):
        """ Plot sound pressure level in the environment at t = 0 """
        fig,ax = plt.subplots()
        surface = ax.pcolormesh(self.X, self.Y, 20*np.log10(np.abs(self.p)/20e-6), vmin = clim[0],vmax = clim[1])
        fig.colorbar(surface)
        plt.axis('equal')
        plt.title('Sound Pressure Level')
        
    def displacement(self,dotres,scale):
        """Plot displacement field as a bunch of dots"""
        plt.figure()
        xmin = min(self.x)
        xmax = max(self.x)
        ymin = min(self.y)
        ymax = max(self.y)
        dotx = np.linspace(xmin, xmax, np.int(np.floor((xmax-xmin)*dotres)))
        doty = np.linspace(ymin, ymax, np.int(np.floor((ymax-ymin)*dotres)))
        dotX, dotY = np.meshgrid(dotx,doty)
        interpfun_rx = RectBivariateSpline(self.x,self.y,np.real(self.gradpx))
        interpfun_ry = RectBivariateSpline(self.x,self.y,np.real(self.gradpy))
        gradpx_interp = interpfun_rx(dotx,doty)*scale
        gradpy_interp = interpfun_ry(dotx,doty)*scale
        plt.plot(dotX+gradpx_interp,dotY+gradpy_interp,'k.')
        plt.axis('equal')
        plt.extent = (self.X.min(), self.X.max(), self.Y.min(), self.Y.max())
        plt.title('Displacement')
        
    def addSource(self, source):
        """ Adds a source to the environment
        INPUTS
        ------
        source - object - currently either a monopole or a plane wave
        """
        # add the source to the environments source list
        self.sourceList.append(source)
        
        # add the pressure to internal 
        pAdd = source.func(self.freq, self.X, self.Y, self.c, self.rho)
        self.p = self.p+pAdd
        
        # calculate complex displacement 
        gradp = np.gradient(self.p)
        self.gradpy = gradp[0]
        self.gradpx = gradp[1]
        
    def animatePressure(self, clim):
        """Displays a looping animation of the pressure field
        INPUTS 
        ------
        clim: 2-array colormap limits [minimum, maximum]
        """
        fig,ax = plt.subplots()
        ax.extent = (self.X.min(), self.X.max(), self.Y.min(), self.Y.max())
        surface = ax.pcolormesh(self.X,self.Y,np.real(self.p),vmin=clim[0],vmax=clim[1]);
        fig.colorbar(surface)
        plt.axis('equal')
        plt.close() # close the figure to prevent static and html5 video figures

        def init():
            """Initialisation of animation plot """
            surface.set_array(np.real(self.p))
            return (surface)

        def animate(i):
            """Called each frame 
            multiplies the array by a phasor to give harmonic output""" 
            surface.set_array(np.array(np.real(self.p*np.exp(-1j*(i/100)*2*np.pi)))[:-1, :-1].ravel())
            return (surface,)
        return HTML(animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=False, repeat=True).to_html5_video())
    

    def animateDisplacement(self,dotres=20,scale=1):
        """Displays a looping animation of the displacement field
        
        INPUTS 
        ------
        dotres - scalar dot resolution per metre
        scale - scalar multiplies pressure gradient to give displacement
        """ 
        xmin = min(self.x)
        xmax = max(self.x)
        ymin = min(self.y)
        ymax = max(self.y)
        dotx = np.linspace(xmin, xmax, np.int(np.floor((xmax-xmin)*dotres)))
        doty = np.linspace(ymin, ymax, np.int(np.floor((ymax-ymin)*dotres)))
        dotX, dotY = np.meshgrid(dotx,doty)
        interpfun_rx = RectBivariateSpline(self.x,self.y,np.real(self.gradpx))
        interpfun_ry = RectBivariateSpline(self.x,self.y,np.real(self.gradpy))
        interpfun_ix = RectBivariateSpline(self.x,self.y,np.imag(self.gradpx))
        interpfun_iy = RectBivariateSpline(self.x,self.y,np.imag(self.gradpy))
        gradp_rx_interp = interpfun_rx(dotx,doty)*scale
        gradp_ry_interp = interpfun_ry(dotx,doty)*scale
        gradp_ix_interp = interpfun_ix(dotx,doty)*scale
        gradp_iy_interp = interpfun_iy(dotx,doty)*scale
        gradpx_interp = gradp_rx_interp+1j*gradp_ix_interp
        gradpy_interp = gradp_ry_interp+1j*gradp_iy_interp
        fig,ax = plt.subplots()
        ax.extent = (self.X.min(), self.X.max(), self.Y.min(), self.Y.max())
        dotplot, = ax.plot([],[],'k.');
        ax.set_xlim((xmin,xmax))
        ax.set_ylim((ymin,ymax))   
        ax.set_aspect('equal')
        plt.close() # close the figure to prevent static and html5 video figures

        def init():
            """Initialisation of animation plot """
            dotplot.set_data(dotX+np.real(gradpx_interp),dotY-np.real(gradpy_interp))
            return (dotplot)

        def animate(i):
            """Called each frame 
            multiplies the array by a phasor to give harmonic output""" 
            dotplot.set_data(dotX+np.real(gradpx_interp*np.exp(-1j*(i/100)*2*np.pi)),dotY-np.real(gradpy_interp*np.exp(-1j*(i/100)*2*np.pi)))
            return (dotplot,)

        return HTML(animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=False, repeat=True).to_html5_video())
    

    def animateDual(self,clim=[-1,1],dotres=20,scale=1):
        """Draw an animated display of pressure and displacement on the same axes

        INPUTS
        ------
        clim: 

        """
        xmin = min(self.x)
        xmax = max(self.x)
        ymin = min(self.y)
        ymax = max(self.y)
        dotx = np.linspace(xmin, xmax, np.int(np.floor((xmax-xmin)*dotres)))
        doty = np.linspace(ymin, ymax, np.int(np.floor((ymax-ymin)*dotres)))
        dotX, dotY = np.meshgrid(dotx,doty)
        interpfun_rx = RectBivariateSpline(self.x,self.y,np.real(self.gradpx))
        interpfun_ry = RectBivariateSpline(self.x,self.y,np.real(self.gradpy))
        interpfun_ix = RectBivariateSpline(self.x,self.y,np.imag(self.gradpx))
        interpfun_iy = RectBivariateSpline(self.x,self.y,np.imag(self.gradpy))
        gradp_rx_interp = interpfun_rx(dotx,doty)*scale
        gradp_ry_interp = interpfun_ry(dotx,doty)*scale
        gradp_ix_interp = interpfun_ix(dotx,doty)*scale
        gradp_iy_interp = interpfun_iy(dotx,doty)*scale
        gradpx_interp = gradp_rx_interp+1j*gradp_ix_interp
        gradpy_interp = gradp_ry_interp+1j*gradp_iy_interp
        fig,ax = plt.subplots()
        ax.extent = (self.X.min(), self.X.max(), self.Y.min(), self.Y.max())
        surface = ax.pcolormesh(self.X,self.Y,np.real(self.p),vmin=clim[0],vmax=clim[1]);
        fig.colorbar(surface)
        dotplot, = ax.plot([],[],'k.');
        sel_white = np.random.choice([True,False],dotX.shape,p = [0.1,0.9])
        dotplot_w, = ax.plot([],[],'w.');
        ax.set_xlim((xmin,xmax))
        ax.set_ylim((ymin,ymax))   
        ax.set_aspect('equal')
        plt.close() # close the figure to prevent static and html5 video figures

        def init():
            """Initialisation of animation plot """
            surface.set_array(np.real(self.p))
            dotplot.set_data(dotX+np.real(gradpx_interp),dotY-np.real(gradpy_interp))
            dotplot_w.set_data(dotX[sel_white]+np.real(gradpx_interp[sel_white]),dotY[sel_white]-np.real(gradpy_interp[sel_white]))
            return (dotplot,surface,dotplot_w)

        def animate(i):
            """Called each frame 
            multiplies the array by a phasor to give harmonic output""" 
            dotplot.set_data(dotX+np.real(gradpx_interp*np.exp(-1j*(i/100)*2*np.pi)),dotY-np.real(gradpy_interp*np.exp(-1j*(i/100)*2*np.pi)))
            dotplot_w.set_data(dotX[sel_white]+np.real(gradpx_interp[sel_white]*np.exp(-1j*(i/100)*2*np.pi)),dotY[sel_white]-np.real(gradpy_interp[sel_white]*np.exp(-1j*(i/100)*2*np.pi)))
            surface.set_array(np.array(np.real(self.p*np.exp(-1j*(i/100)*2*np.pi)))[:-1, :-1].ravel())
            return (dotplot, dotplot_w, surface,)

        return HTML(animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=False, repeat=True).to_html5_video())
  
  
class monopole(object):
    """Acoustic monopole source"""
    def __init__(self, coords, Q, phase):
        """Define a monopole source 
        INPUTS
        ------
        coords: 2-array [x,y]
        Q: scalar source strength
        phase: scalar phase (0<phase<2*pi)
        """
        self.x, self.y = coords
        self.Q = Q
        self.phase = phase
        # self.func is a function which takes freq, x and y coordinates of points, speed of sound and density from the
        # environment, to calculate pressure amplitude
        self.func = lambda freq, x, y, c, rho: ( 
                ((1j*2*np.pi*freq/c)*rho*Q)/(4*np.pi*np.sqrt((x-self.x)**2+(y-self.y)**2)))*np.exp(1j*((2*np.pi*freq/c*(np.sqrt((x-self.x)**2+(y-self.y)**2)))+phase))
        
class planeWave(object):
    """Acoustic plane wave source"""
    def __init__(self, direction, phase):
        """Define a plane wave source 
        INPUTS
        ------
        direction: 2-array direction vector
        phase: scalar phase (0<phase<2*pi)"""
        theta = np.angle(direction[0]+1j*direction[1])
        self.func = lambda freq, x, y, c, rho: (
                np.exp(1j*((2*np.pi*freq*np.cos(theta)/c)*x+(2*np.pi*freq*np.sin(theta)/c)*y)))
 
```

## Example 2: Source Addition

Sources can be added together in an environment

```python
E2 = env(xlim=[-1, 1], ylim=[-1, 1], resolution=200, freq = 1000, c=343, rho=1.225)
P = planeWave([1,0.5],0)
M = monopole([0,0], Q = 0.06, phase=0)
E2.addSource(M)
E2.addSource(P)
E2.animateDual(clim=[-2,2],scale=0.2)
```

## Example 3 - Phase shift

```python
E3 = env(xlim=[-2, 2], ylim=[-2, 2], resolution=100, freq = 1000, c=343, rho=1.225)
M1 = monopole([0, 0], Q=0.03, phase=0)
M2 = monopole([0, 0.1], Q=0.03, phase=np.pi)
E3.addSource(M)
E3.addSource(M2)
E3.animatePressure([-1,1])
```

## Example 4: Arrays 
This demonstration shows the creation of an array of sources in a loop.

```python
E4 = env(xlim=[-1, 1], ylim=[-0.2, 1.8], resolution=200, freq=1000, c=343, rho=1.225)
n = 10
xcoords = np.linspace(-0.3, 0.3, n)
ycoords = np.sin
phases = np.linspace(0, np.pi, n)
for xCoord, theta in zip(xcoords, phases):
    M = monopole([xCoord, 0], 0.01, theta)
    E4.addSource(M)
E4.pressure([-1,1])
E4.spl([50,100])
E4.animateDual()
```
