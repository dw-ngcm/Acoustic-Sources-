The following notebook contains a minimum working example of an acoustic environment and a monopole source class

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
class env(object):
    """ 2D environment

    METHODS
    -------
    __init__ : initialise internal variables
    pressure : plot pressure amplitude
    spl : plot sound pressure level """

    def __init__(self, xlim, ylim, resolution, c=343, rho=1.225):
        """ Initialise internal variables
        INPUTS
        ------
        x = 2-tuple (xmin, xmax)
        y = 2-tuple (ymin, ymax)
        resolution = scalar number of points per linear metre
        c = scalar speed of sound

        """
        assert len(list(xlim)) == 2, 'Usage: xlim = [xmin, xmax]'
        assert len(list(ylim)) == 2, 'Usage: ylim = [ymin, ymax]'
        xmin, xmax = xlim
        ymin, ymax = ylim
        x = np.linspace(xmin, xmax, np.int(np.floor((xmax-xmin)*resolution)))
        y = np.linspace(ymin, ymax, np.int(np.floor((ymax-ymin)*resolution)))
        self.X, self.Y = np.meshgrid(x, y)  # set X and Y coordinate grids
        self.p = np.zeros_like(self.X, dtype=np.complex128)  # pressure grid
        self.sourceList = []
        self.c = c
        self.rho = rho
        
    def pressure(self,clim=(-2,2)):
        """ Plot pressure in the environment at t = 0 """
        plt.figure()
        plt.imshow(np.real(self.p), extent=(self.X.min(), self.X.max(),
                                            self.Y.min(), self.Y.max()))
        plt.colorbar()
        plt.title('Pressure Amplitude (Pa)')
        plt.clim(clim)

    def spl(self):
        """ Plot sound pressure level in the environment at t = 0 """
        plt.figure()
        plt.imshow(20*np.log10(np.abs(self.p)/20e-6),
                   extent=(self.X.min(), self.X.max(),
                   self.Y.min(), self.Y.max()))
        plt.colorbar()
        plt.title('Sound Pressure Level')

    def addSource(self, source):
        """ Adds a source to the environment
        INPUTS
        ------
        source - object of superclass source
            could be a monopole
        """
        self.sourceList.append(source)
        r = np.sqrt((self.X-source.x)**2+(self.Y-source.y)**2)
        pAdd = source.func(r, self.c, self.rho)
        self.p = self.p+pAdd


class monopole(object):
    """ monopole source

    METHODS
    -------
    __init__ : generate source

    """
    def __init__(self, coords, Q, freq, phase):
        self.x, self.y = coords
        self.Q = Q
        self.freq = freq
        self.phase = phase
        self.func = lambda r, c, rho: ( 
                ((1j*2*np.pi*freq/c)*rho*Q)/(4*np.pi*r))*np.exp(1j*((2*np.pi*freq/c*r)+phase))
```

The following example generates a single 1kHz point source at the centre of the view-window

```python
E = env(xlim=[-0.5, 0.5], ylim=[-0.5, 0.5], resolution=200, c=343, rho=1.225)
M = monopole([0, 0], Q=0.03, freq=1000, phase=0)
E.addSource(M)
E.pressure([-1,1])
E.spl()
```

The next demonstration creates an array of sources in a loop.

```python
E = env(xlim=[-0.5, 1], ylim=[-1., 1.], resolution=200, c=343, rho=1.225)
n = 10
xcoords = np.linspace(-0.3, 0.3, n)
ycoords = np.sin
phases = np.linspace(0, np.pi, n)
for xCoord, theta in zip(xcoords, phases):
    M = monopole([xCoord, 0], 0.01, 1000, theta)
    E.addSource(M)
E.pressure([-1,1])
E.spl()
```

Circular arrays can be made similarly

```python
E = env(xlim=[-1., 1.], ylim=[-1., 1.], resolution=200)
n = 3
r = 0.5
angles = np.linspace(0, 2*np.pi, n+1)[0:-1]
xcoords = np.cos(angles)*r
ycoords = np.sin(angles)*r
phases = np.linspace(0, np.pi, n+1)[0:-1]
for xCoord, yCoord, theta in zip(xcoords, ycoords, phases):
    M = monopole([xCoord, yCoord], 0.01, 1000, theta)
    E.addSource(M)
E.pressure([-1,1])
E.spl()
```

Sources of different frequencies create interesting interference patterns. This is not as easy to animate... 

```python
E = env(xlim=[-1., 1.], ylim=[-1., 1.], resolution=200)
n = 3
r = 0.5
M1 = monopole(coords=[0.5,0],Q=0.01,freq=1000,phase=0)
M2 = monopole(coords=[-0.5,0],Q=0.01,freq=1500,phase=0)
E.addSource(M1)
E.addSource(M2)
E.pressure([-0.5,0.5])
E.spl()
```
